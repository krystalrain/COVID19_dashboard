```python
url = 'https://data.cdc.gov/resource/k8wy-p9cg.json'
 
response = requests.get(url)
data = response.text
parsed1 = json.loads(data)
print(json.dumps(parsed1, indent=4))
```

```python
p1= pd.DataFrame(parsed1)
```

```python
p1
```

```python
p1.columns
```

```python
p1=p1.drop(columns= ['data_as_of', 'start_week','end_week','urbanruralcode', 'fipsstate', 'fipscounty', 'fipscode', 'urbanruraldesc', 'footnote'])
```

```python
df = pd.DataFrame(data=p1)
df.dtypes
df=df.astype({'covid_19_deaths_total': 'int32'})
df=df.astype({'state': 'str'})
df=df.astype({'county_name': 'str'})
```

```python
df = df.astype({'non_hispanic_white': 'float', 'non_hispanic_black': 'float', 'non_hispanic_american_indian':'float',  'non_hispanic_asian': 'float', 'non_hispanic_nhopi':'float', 'hispanic':'float', 'other': 'float'})
```

```python
df.head()
```

```python
df.drop(df.loc[df['indicator'] == 'Distribution of all-cause deaths (%)'].index, inplace=True)
```

```python
df.drop(df.loc[df['indicator'] == 'Distribution of population (%)'].index, inplace=True)
```

```python
grouped = df.groupby(['state','county_name'])['covid_19_deaths_total'].sum()
grouped
a=grouped.groupby(level='state').nlargest(3).reset_index(level=0, drop=True)
```

# Top 10 Covid-Deaths states

```python

a1 = df.groupby(['state'])['covid_19_deaths_total'].sum()
a1 =pd.DataFrame(a1)
a1=a1.sort_values(["covid_19_deaths_total"], ascending=False).head(10)
```

```python
a1.plot(kind='bar',figsize=(10,5))
```

### The top 3 states that have the highest number of the deaths:
### 1. California
### 2. Florida
### 3. Illinois
### Rest all have more or less same number of deaths, but California requires the major attention.

# Covid-Deaths as per State+County

```python
grouped = df.groupby(['state','county_name'])['covid_19_deaths_total'].sum()
grouped
a=grouped.groupby(level='state').nlargest(3).reset_index(level=0, drop=True)
a=pd.DataFrame(a)
```

```python
a.pivot_table(index='county_name', columns='state', values='covid_19_deaths_total', aggfunc=sum)
a.plot(kind='bar',stacked=True, figsize=(20,10))
```

### The total number of death counts were the highest and mmost affected counties were
### 1. Los Angeles County in California
### 2. Cook County in Illinois
### 3. Maricopa County in Arizona
